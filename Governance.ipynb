1. Governance Score by Company

```python
import pandas as pd
import plotly.express as px

df = pd.read_csv("company_esg_financial_dataset.csv")

df = df.dropna(subset=["CompanyName","Year","ESG_Governance"])

df["ESG_Governance"] = pd.to_numeric(df["ESG_Governance"], errors="coerce")

latest = df.sort_values(["CompanyName","Year"]).groupby("CompanyName").tail(1)

def risk_level(score):
    if score >= 75:
        return "Low Risk"
    elif score >= 55:
        return "Medium Risk"
    else:
        return "High Risk"

latest["Risk Level"] = latest["ESG_Governance"].apply(risk_level)

fig = px.bar(
    latest.sort_values("ESG_Governance", ascending=False),
    x="CompanyName",
    y="ESG_Governance",
    color="Risk Level",
    title="Governance Score by Company"
)

fig.show()
```

2. Compliance Heatmap

```python
import pandas as pd
import plotly.express as px

df = pd.read_csv("company_esg_financial_dataset.csv")

df = df.dropna(subset=["CompanyName","Year","ESG_Governance","ESG_Overall","ProfitMargin"])

latest = df.sort_values(["CompanyName","Year"]).groupby("CompanyName").tail(1)

comp = latest[["CompanyName","ESG_Governance","ESG_Overall","ProfitMargin"]].copy()

comp["Board Compliance"] = (comp["ESG_Governance"]*0.9 + comp["ESG_Overall"]*0.1)
comp["Audit Compliance"] = (comp["ESG_Governance"]*0.75 + comp["ESG_Overall"]*0.25)
comp["Regulatory Compliance"] = (comp["ESG_Governance"]*0.7 + (comp["ProfitMargin"]+20))

heatmap_data = comp.set_index("CompanyName")[[
    "Board Compliance",
    "Audit Compliance",
    "Regulatory Compliance"
]]

fig = px.imshow(
    heatmap_data,
    text_auto=True,
    aspect="auto",
    title="Compliance Heatmap"
)

fig.show()
```

3. Governance Risk Distribution

```python
import pandas as pd
import plotly.express as px

df = pd.read_csv("company_esg_financial_dataset.csv")

df = df.dropna(subset=["CompanyName","Year","ESG_Governance"])

latest = df.sort_values(["CompanyName","Year"]).groupby("CompanyName").tail(1)

def risk_level(score):
    if score >= 75:
        return "Low Risk"
    elif score >= 55:
        return "Medium Risk"
    else:
        return "High Risk"

latest["Risk Level"] = latest["ESG_Governance"].apply(risk_level)

counts = latest["Risk Level"].value_counts().reset_index()
counts.columns = ["Risk Level", "Count"]

fig = px.pie(
    counts,
    names="Risk Level",
    values="Count",
    title="Governance Risk Distribution"
)

fig.show()
```
