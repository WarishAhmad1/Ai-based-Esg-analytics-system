1. Customer Satisfaction & Engagement

```python
import pandas as pd
import plotly.express as px

df = pd.read_csv("company_esg_financial_dataset.csv")

df = df.dropna(subset=["CompanyName","Year","ESG_Social","ESG_Governance"])

latest = df.sort_values(["CompanyName","Year"]).groupby("CompanyName").tail(1)

latest["Customer Satisfaction"] = latest["ESG_Social"]
latest["Engagement"] = latest["ESG_Social"]*0.7 + latest["ESG_Governance"]*0.3

fig = px.bar(
    latest,
    x="CompanyName",
    y=["Customer Satisfaction", "Engagement"],
    barmode="group",
    title="Customer Satisfaction & Engagement"
)

fig.show()
```

2. Employee Turnover

```python
import pandas as pd
import plotly.express as px

df = pd.read_csv("company_esg_financial_dataset.csv")

df = df.dropna(subset=["CompanyName","Year","ESG_Social"])

latest = df.sort_values(["CompanyName","Year"]).groupby("CompanyName").tail(1)

latest["Employee Turnover (%)"] = (35 - latest["ESG_Social"]*0.3).clip(5,35)

fig = px.bar(
    latest,
    x="CompanyName",
    y="Employee Turnover (%)",
    color="Employee Turnover (%)",
    title="Employee Turnover by Company"
)

fig.show()
```

3. Diversity & Inclusion

```python
import pandas as pd
import plotly.express as px

df = pd.read_csv("company_esg_financial_dataset.csv")

df = df.dropna(subset=["CompanyName","Year","ESG_Social","ESG_Governance"])

latest = df.sort_values(["CompanyName","Year"]).groupby("CompanyName").tail(1)

latest["Diversity (%)"] = (latest["ESG_Social"]*0.6).clip(20,70)
latest["Inclusion (%)"] = (latest["ESG_Governance"]*0.5).clip(20,70)

melted = latest.melt(
    id_vars="CompanyName",
    value_vars=["Diversity (%)","Inclusion (%)"],
    var_name="Metric",
    value_name="Value"
)

fig = px.bar(
    melted,
    x="CompanyName",
    y="Value",
    color="Metric",
    barmode="stack",
    title="Diversity & Inclusion"
)

fig.show()
```
