1. ESG Score Trend Over Time

```python
import pandas as pd
import plotly.express as px

df = pd.read_csv("company_esg_financial_dataset.csv")

df = df.dropna(subset=["Year","ESG_Overall"])

df["Year"] = pd.to_numeric(df["Year"], errors="coerce")
df["ESG_Overall"] = pd.to_numeric(df["ESG_Overall"], errors="coerce")

df = df.dropna()

trend = df.groupby("Year", as_index=False)["ESG_Overall"].mean()

fig = px.line(
    trend,
    x="Year",
    y="ESG_Overall",
    markers=True,
    title="Average ESG Score Trend"
)

fig.show()
```

2. Company Comparison

```python
import pandas as pd
import plotly.express as px

df = pd.read_csv("company_esg_financial_dataset.csv")

df = df.dropna(subset=["CompanyName","Year","ESG_Overall"])

df["Year"] = pd.to_numeric(df["Year"], errors="coerce")
df["ESG_Overall"] = pd.to_numeric(df["ESG_Overall"], errors="coerce")

df = df.dropna()

company_trend = df.groupby(["Year","CompanyName"], as_index=False)["ESG_Overall"].mean()

latest = company_trend.sort_values(["CompanyName","Year"]).groupby("CompanyName").tail(1)

top_companies = latest.nlargest(8, "ESG_Overall")["CompanyName"]

filtered = company_trend[company_trend["CompanyName"].isin(top_companies)]

fig = px.line(
    filtered,
    x="Year",
    y="ESG_Overall",
    color="CompanyName",
    markers=True,
    title="Top Companies ESG Comparison"
)

fig.show()
```

3. Growth Pattern

```python
import pandas as pd
import plotly.express as px

df = pd.read_csv("company_esg_financial_dataset.csv")

df = df.dropna(subset=["CompanyName","Year","ESG_Overall"])

df["Year"] = pd.to_numeric(df["Year"], errors="coerce")
df["ESG_Overall"] = pd.to_numeric(df["ESG_Overall"], errors="coerce")

df = df.dropna()

company_trend = df.groupby(["Year","CompanyName"], as_index=False)["ESG_Overall"].mean()

latest = company_trend.sort_values(["CompanyName","Year"]).groupby("CompanyName").tail(1)

top_companies = latest.nlargest(8, "ESG_Overall")["CompanyName"]

filtered = company_trend[company_trend["CompanyName"].isin(top_companies)]

fig = px.area(
    filtered,
    x="Year",
    y="ESG_Overall",
    color="CompanyName",
    title="ESG Growth Pattern"
)

fig.show()
```
